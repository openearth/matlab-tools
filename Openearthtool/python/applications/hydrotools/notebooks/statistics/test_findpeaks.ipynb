In this Notebook the findpeaks.py script is illustrated with a time series of daily discharge. 

This function finds peaks in a time series, e.g. a hydrograph. A peak is a data 
sample that is larger than its two neighboring samples AND the largest sample 
within a moving time window.

Three methods are available: 
peaks over threshold, annual maxima, selection of n peaks

```python
%matplotlib inline
from hydrotools.statistics import findpeaks as fp
import pandas as pd
import matplotlib.pyplot as plt
```

Let's load some sample data
try to load the data using a relative path 
OR edit the directories to subdirectory data to find the hydrograph data

```python
try:
    fname = r"data\hydrograph.xls"
    sample = pd.read_excel(fname,index_col=0)
except:     
    OETdir = "d:\\scripts\\OET\\"
    DATAdir = "python\\applications\\hydrotools\\notebooks\\statistics\\data\\"
    fname = "hydrograph.xls"
    sample = pd.read_excel(OETdir + DATAdir + fname,index_col=0)
```

First, we will find all peaks given a 30 days period between peaks and a minumum value (threshold). The distance betweeen peaks is important in order to be able to treat them as independent events. This yields only the largest peaks

```python
peak_dist = 30
threshold = 200 #m3/s
POT2 = fp.peaks_over_threshold(sample, threshold=threshold, peak_dist=peak_dist)

fig = plt.figure()
ax = fig.add_subplot(111)
sample.plot(ax=ax,style='-b', legend=False)
POT2.plot(ax=ax,style='og',markersize=6, legend=False)
ax.legend(['Discharge','peaks above threshold'],fontsize=10)
fig.show()
```

To find a good value for the threshold we use the find_n_peaks function, with n equal to the number of years of the hydrograph. This function finds the the n largest peaks in the hydrograph.

```python
n_peaks=round(len(sample)/365,0)
peak_dist = 30
POT3 = fp.find_n_peaks(sample, n_peaks=n_peaks, threshold=None, peak_dist=peak_dist)
threshold = POT3.min()

print 'the found threshold = ' + str(round(threshold,2)) + ' m3/s'

fig = plt.figure()
ax = fig.add_subplot(111)
sample.plot(ax=ax,style='-b', legend=False)
POT3.plot(ax=ax,style='og',markersize=6, legend=False)
ax.legend(['Discharge','n peaks equal to no. of years'],fontsize=10)
fig.show()
```

Alternatively, peaks can be selected based on the maximum peak of a hydrological year.

```python
first_month = 'OCT'
peak_dist = 15
AM = fp.annual_max(sample,first_month=first_month,threshold=None,peak_dist=peak_dist)

fig = plt.figure()
ax = fig.add_subplot(111)
sample.plot(ax=ax,style='-b', legend=False)
AM.plot(ax=ax,style='og',markersize=6, legend=False)
ax.legend(['Discharge','annual maxima peaks'],fontsize=10)
fig.show()
```

The following figure compares the peaks selected with all three method

```python
fig = plt.figure()
ax = fig.add_subplot(111)
sample.plot(ax=ax,style='-b', legend=False)
AM.plot(ax=ax,style='*y',markersize=10, legend=False)
POT2.plot(ax=ax,style='or',markersize=10, legend=False)
POT3.plot(ax=ax,style='og',markersize=5, legend=False)
ax.legend(['Discharge','AM','peaks over threshold method','peaks equal to no. of years'],fontsize=10)
fig.show()
```
