Let's first do the boring (but important) part. We need a function that interpolates point values on a regular grid. An easy one is inverse distance. With inverse distance we reduce the weight of a station with a power of its distance to the interpolation point. Easy to apply for a full grid.

Furthermore, we make a function that transforms a shapefile with the island into a mask, so that we can maks out ocean values later on.

```python
def inverse_distance(xcoords, ycoords, values, xi, yi, distance_weight):
    """
    This function interpolates the values in xcoords, ycoords to regular grid xi, yi using a power weight in distance_weight
    """
    # first determine the weights per station
    weights = np.empty((len(values), xi.shape[0], xi.shape[1]))
    for n, (x, y) in enumerate(zip(xcoords, ycoords)):
        dist = ((xi-x)**2 + (yi-y)**2)**0.5  # we apply pythagoras' law to each grid point
        weight = 1./(dist**distance_weight)
        weights[n, :, :] = weight
    
    # normalize weights per cell so that they add up to unity, and compute the contribution of each value
    interpolation = np.zeros(xi.shape)  # make a zero matrix of the size of xi
    for n, value in enumerate(values):
        cur_weight = weights[n, :, :]/np.sum(weights,axis=0)
        # now we have the weights, interpolation is a piece of cake!
        interpolation += cur_weight*value
    
    # return the interpolated values
    return interpolation
    
def mask_shape(lyr, xcount, ycount, xmin, ymax, pixel_width, pixel_height):
    # create memory target raster
    ds = gdal.GetDriverByName('MEM').Create('', xcount, ycount, gdal.GDT_Byte)
    ds.SetGeoTransform((
        xmin, pixel_width, 0,
        ymax, 0, -pixel_height,
    ))
    
    # rasterize zone polygon to raster
    gdal.RasterizeLayer(ds, [1], lyr, burn_values=[1])
    # create smaller bits of work if necessary
    # read raster as arrays and read Fill Value
    bandmask = ds.GetRasterBand(1) 
    datamask = bandmask.ReadAsArray(0, 0, xcount, ycount) # .astype(numpy.float32) 
    return datamask
```

Awesome! Our function is ready, now let's make some fun! First we'll read a list of stations. We assume that the last 7 are the ones that monitor temperature

```python
stations_file = r'data\Stations_Curacao.csv'
stations = pd.read_csv(stations_file)
stations
stations_select = stations[-7:]
stations_select
xcoords = stations_select['X']
ycoords = stations_select['Y']
print xcoords, ycoords
```

```python
# now we make some fake values for temperature
values = [29, 30, 32, 34, 30.5, 30.9, 32.6]
```

now we have station data at some locations and we understand which value belongs to which location, let's apply the interpolation function

```python
# we need a nice grid
xax = np.arange(-69.25, -68.65, 0.001)
yax = np.arange(11.95, 12.45, 0.001)
xi, yi = np.meshgrid(xax, yax)
xi
interpolation = inverse_distance(xcoords, ycoords, values, xi, yi, 2)

# now prepare the mask and remove values outside the mask
shapefile = r'd:\courses\python_matlabusers\ipynb\curacao_shape\CUW_adm0.shp'
driver = ogr.GetDriverByName('ESRI Shapefile') 
shp = driver.Open(shapefile, 1)
lyr = shp.GetLayer() 

mask = np.flipud(mask_shape(lyr, len(xax), len(yax), xax[0], yax[-1], 0.001, 0.001))
interpolation[mask==0] = -999

interpolation_masked = np.ma.masked_equal(interpolation, -999)


```

```python
plt.imshow(interpolation_masked)
plt.colorbar()
```

This is a very ugly graph with no geographical reference whatsoever, the island even looks up-side-down because of those silly (UUUGGGGHHH) GIS folks who always turn axes around! let's make a nicer plot in cartopy (sorry, thid was a fast do-e, so the plot could be much nicer still, but time was limited).

```python
plt.figure(figsize=(15,15))
ax = plt.axes(projection=ccrs.PlateCarree())

# load in a nice tiler, you can also use a different one, see cartopy manual
tiler = GoogleTiles()

# get hold of the coastlines for that area.
extent = (xi.min(), xi.max(), yi.min(), yi.max())
ax.set_extent(extent)
ax.add_image(tiler, 11)

norm = np.arange(25, 35, 0.5)
yyyy = 2015
mm = 7
dd = 31
HH = 12
file_template = 'temperature_{:d}{:d}{:d}{:02d}.png'
p = ax.contourf(xi, yi, interpolation_masked, norm,
            transform=ccrs.PlateCarree(),
            cmap=cm.jet, extend='both',
            alpha=0.5)
p2 = ax.contour(xi, yi, interpolation_masked, norm,
            transform=ccrs.PlateCarree(),
            colors='k', extend='both')  # colors='#AAAAAA'
plt.clabel(p2, fmt='%d', colors='k', fontsize=14)

cax = plt.axes([0.8,0.25,0.01,0.5])
cbar = plt.colorbar(p, extend='both', cax=cax)
cbar.set_label('oC', fontsize=16)
#plt.show()
plt.savefig(file_template.format(yyyy, mm, dd, HH), dpi=300)
```
