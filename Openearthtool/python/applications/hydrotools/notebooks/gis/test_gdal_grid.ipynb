Gridding irregularly spaced data over a very large grid is sometimes difficult due to memory issues. In this notebook, the gdal_grid function is demonstrated. gdal_grid outputs a GeoTIFF file directly rather than keeping everything in memory. It can interpolate a list of point data (i.e. x coordinates, y coordinates and a set of values) into a grid. This can be used for instance to interpolate rainfall data or other irregularly spaced data into a high resolution grid.

```python
# import packages
from hydrotools import gis
```

As an example let's make a function that calculates some data based on x, y values as input. We use the function used in scipy.interpolate.griddata as example. And let's see what it looks like if we evaluate this function over a nice grid.

```python
def func(x, y):
    return x*(1-x)*np.cos(4*np.pi*x) * np.sin(4*np.pi*y**2)**2

# try the function on a grid
yax = np.linspace(0, 1, 1000)  # make some yax
xax = np.linspace(0, 1, 1000)  # make some xax

xi, yi = np.meshgrid(xax, yax)
some_gridded_data = func(xi, yi)

# let's plot
plt.imshow(some_gridded_data)
```

Let's now assume we only know the values in the grid above in a few (say, 20) points and let's interpolate these with gdal_grid

```python
nr_of_samples = 20

points = np.float32(np.random.rand(nr_of_samples, 2))  # these are the y, x locations at some random locations
values = func(points[:,0], points[:,1])  # these are the values at the y, x locations

# now let's interpolate with nearest neighbour. After interpolation, we will read back the generated file
filename = 'test.tif'
gis.gdal_grid(filename, points, values, (yax, xax), method='nearest', zlib=True)

x, y, nearest, fill = gis.gdal_readmap(filename, 'GTiff')
plt.imshow(nearest)
```

with more samples, the results will look more as in the "real" grid

```python
points1 = np.float32(np.random.rand(1000, 2))  # these are the y, x locations at some random locations
values1 = func(points1[:,0], points1[:,1])  # these are the values at the y, x locations
points2 = np.float32(np.random.rand(10000, 2))  # these are the y, x locations at some random locations
values2 = func(points2[:,0], points2[:,1])  # these are the values at the y, x locations

# now let's interpolate with nearest neighbour. After interpolation, we will read back the generated file
filename = 'test.tif'
gis.gdal_grid(filename, points1, values1, (yax, xax), method='nearest', zlib=True)

x, y, nearest1, fill = gis.gdal_readmap(filename, 'GTiff')

gis.gdal_grid(filename, points2, values2, (yax, xax), method='nearest', zlib=True)

x, y, nearest2, fill = gis.gdal_readmap(filename, 'GTiff')
plt.subplot(121)

plt.imshow(nearest1)
plt.title('1,000 samples')

plt.subplot(122)
plt.imshow(nearest2)
plt.title('10,000 samples')
```

Let's see what happens if we use chunksizes of 300 x 300 pixels. with 1000 x 1000 grid sizes, this is really not needed, but with 50,000 x 50,000 it is recommended to use this! You'll see some edge effects where you move from one chunk to the next. We use linear interpolation to really show this exaggerated.

```python
points3 = np.float32(np.random.rand(1000, 2))  # these are the y, x locations at some random locations
values3 = func(points3[:,0], points3[:,1])  # these are the values at the y, x locations

# now let's interpolate with nearest neighbour. After interpolation, we will read back the generated file
filename = 'test.tif'
gis.gdal_grid(filename, points3, values3, (yax, xax), method='linear',
              chunksizes=(300, 300), zlib=True)

x, y, nearest3, fill = gis.gdal_readmap(filename, 'GTiff')
nearest3 = np.ma.masked_equal(nearest3, fill)  # mask out missing areas

plt.imshow(nearest3)
plt.colorbar()
plt.title('Some nasty edges every 300 pixels')
```

The edge effects are due to the fact that samples from outside each chunk are not used, although they may contain the nearest neighbour, or are needed to complete a triangulation if linear interpolation is used. So Let's introduce some overlap.

```python
values4 = func(points3[:,0], points3[:,1])  # these are the values at the y, x locations

# now let's interpolate with nearest neighbour. After interpolation, we will read back the generated file
filename = 'test.tif'
gis.gdal_grid(filename, points3, values4, (yax, xax), method='linear',
              chunksizes=(300, 300), chunk_overlap=(100, 100), zlib=True)

x, y, nearest4, fill = gis.gdal_readmap(filename, 'GTiff')
nearest4 = np.ma.masked_equal(nearest4, fill)  # mask out missing areas

plt.imshow(nearest4)
plt.colorbar()
plt.title('Only missings at the edge of the complete figure')
```
