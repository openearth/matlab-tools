```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 10))
wms = r'https://wci.earth2observe.eu/thredds/wms/jrc/wrr1-daily-agg.nc'
ax = plt.axes(projection=ccrs.Robinson())  # ccrs.InterruptedGoodeHomolosine()
ax.coastlines()
kwargs = {
'colorscalerange': '0, 0.00001',
'time': '1979-01-01T00:00:00.000Z',
'abovemaxcolor': '0xFFFFFF'
}
ax.add_wms(wms=wms,
           layers=['ECanop'],
           
           wms_kwargs=kwargs)


plt.show()



```
