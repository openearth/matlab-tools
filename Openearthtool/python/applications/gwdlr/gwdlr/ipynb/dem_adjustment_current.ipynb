```python
import numpy
from termcolor import colored

#DEM Iteration
Zlist = [21.0, 14.0, 16.0, 16.0, 19.0, 21.0, 15.0, 18.0, 11.0, 9.0]
fill_value = -10000   # ocean cell value
Z = numpy.array(Zlist)
print Z
print "======================================================="
#plot(Z)

Z_intval = 1.
w_ups = 10
w_dws = 1

SUM_LMOD = np.zeros(len(Z))
print "SUM_LMOD=", SUM_LMOD

Z_FINAL = np.zeros(10)
print 'Z_FINAL=', Z_FINAL

Z_CHOSEN = np.zeros(10)
print 'Z_CHOSEN=', Z_CHOSEN

for i, z in enumerate(Z[1:-1]):
    if z != fill_value:
        "=======================start==========================="
        print colored("Index!=", 'green'), colored(i, 'green')        
        
        #Error Downstream
        print colored("Check Error Downstream...", 'blue')
        Z_min = Z[i]
        Z_max = z
        print "Z_min =", Z_min
        print "Z_max =", Z_max
        
        # If there are no errors, do nothing and continue ...
        if Z_max <= Z_min :
            print "Z_next <= Z_curr, do nothing, continue..."
            print "--------------------------"
        
        if Z_max > Z_min:
            Z_var = arange(Z_min, Z_max + 1, Z_intval)
            #Z_var_dws = Z_varr_dws[:-1] # Delete -1.0 -- ask Hessel
            print colored("Error Downstream!!!", 'blue')
            print "---------------------------------"
            
            # Define range of Z_variations
            print "RANGE OF Z_VARIATIONS:"
            print "Z_var =", Z_var
            print "  "
            
            # Calculate downstream_modifications:
                 
            #first introduce temporary arrays for listing all Nmod_downtream
            Lmod_dws_array = np.zeros(len(Z_var))
            
            #introduce Z_current for each i index
            Z_current = np.zeros(len(Z_var))
    
            for dws, Z_curr in enumerate(Z_var):
                
                print "Calculate downstream_modifications...."
                print "dws =", dws
                print "Z_min +",dws,"=", Z_curr
                    # Z_min+ first count from Zmin to Zmax with Z_interval
                print "z_more_downstream=", Z[i:]
                Hmod_dws = np.sum(np.maximum((Z[i:] - Z_var[dws]), 0))
                #print "Hmod_dws=", Hmod_dws
                Lmod_dws = Hmod_dws * w_dws
                print "Lmod_dws_",dws,"=",Lmod_dws
                Lmod_dws_array[dws] = Lmod_dws
                print "Lmod_dws_array_",dws,"=", Lmod_dws_array
                
                #Get Z_curr values
                Z_current = Z_curr
                print 'Z_current_dws=', Z_current

            # Return the index to the first index
            # Number_modification_dowstream for original index:
            Lmod_index_dws = Lmod_dws_array
            print "For i =", i
            print "LMOD_INDEX_DWS =", Lmod_index_dws    
            print "--------------------------"
        
                        
            # Calculate upstream_modifications:               
                
            #first introduce temporary arrays for listing all Nmod_upstream
            Lmod_ups_array = np.zeros(len(Z_var))
            
            for ups, Z_curr in enumerate(Z_var):
                print "Calculate upstream_modifications...."
                print "ups =", ups
                print "Z_min +",ups,"=", Z_curr
                    # Z_min+ first count from Zmin to Zmax with Z_interval
                print "z_more_downstream=", Z[:i]
                Hmod_ups = np.sum(np.maximum(Z_curr - Z[:i + 1], 0))
                L = (Z_curr - Z[:i + 1])
                print "Hmod_upskuuu=", L
                #print "Hmod_dws=", Hmod_dws
                Lmod_ups = Hmod_ups * w_ups
                print "Lmod_ups_",ups,"=",Lmod_ups
                Lmod_ups_array[ups] = Lmod_ups
                print "Lmod_ups_array_",ups,"=", Lmod_ups_array
                
                #Get Z_curr values
                Z_current = Z_curr
                print 'Z_current_ups=', Z_current

                # Return the index to the first index
                # Number_modification_dowstream for original index:
            Lmod_index_ups = Lmod_ups_array
            print "For i =", i
            print "LMOD_INDEX_UPS =", Lmod_index_ups    
            print "--------------------------"
        
        if Z_max > Z_min :
        
            #combine Lmod_index downstream and upstream

            from itertools import izip_longest

            print "LMOD_INDEX_DWS =", Lmod_index_dws
            print "LMOD_INDEX_UPS =", Lmod_index_ups

            R = [x+y for x,y in izip_longest(Lmod_index_dws, Lmod_index_ups, fillvalue=0)]
            print "TOTAL ERRORS", R

            MIN_ERRORS = min(R)
            print 'MIN_ERRORS=', MIN_ERRORS

            MIN_ERRORS_INDEX = R.index(min(R))
            print "MIN_ERRORS_INDEX=", MIN_ERRORS_INDEX
            
            #print Z_chosen - use similar procedure as written above
            Z_CHOSEN[i] = Z_current
            print 'Z_CHOSEN=', Z_CHOSEN            

            Z_FINAL[i] = MIN_ERRORS_INDEX
            print "SELECTED MODIFICATION FOR INDEX-",i,"=", Z_FINAL
            
        else:
            pass
        
    else:
        print 'Ocean cell found'
        
#Return selected modification for original Z_array...
print 'Z_CHOSEN=', Z_CHOSEN

# Introduce modified elevation as new variable (Z_MODIFIED)
# and compare the differences with the original elevation (Z)

Z_MODIFIED = Z
print 'Z_ORIGINAL=', Z_MODIFIED

for index, z_value in enumerate(Z_CHOSEN):
    
    if z_value > 0:
        print 'z_index',index,'=', z_value
        Z_MODIFIED[index] = z_value
        
print 'Z_MODIFIED-NEW', Z_MODIFIED

# create index for plotting
import matplotlib.pylab as p

import numpy as np
index_plot = np.arange(1,11)
print 'index_plot=', index_plot

p.plot(index_plot,Zlist,'r')
p.plot(index_plot,Z_MODIFIED,'--k')
p.legend(["Z_original", "Z_modified"])
p.show()

print 'observe the difference between Z_original and Z_modified...'
print 'is there any improvement?'

print "=======================end============================="
```
