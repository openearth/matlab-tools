```python
# <nbformat>3.0</nbformat>

# <markdowncell>

# This file tests a possible stream burning (or smoothing) method to be able to use the elevation values of the DEM for 1D D-Flow FM.
# 
# The idea behind it is that the current DEM would result in very large "mountains" in the river, preventing the flow in 1D to continue. The stream burning algorithm 'smooths' the DEM, by checking if upstream points have a lower elevation than downstream points and then burning the downstream point accordingly.

# <codecell>

# Required libraries
from osgeo import ogr
from collections import defaultdict
import matplotlib.pyplot as plt
import numpy as np

# <codecell>

# Paths

# external disk or local drive
# (1 = ext. disk, 0 = local)
path_option = 0

if path_option == 0:
    
    path_shapefile_points = r"d:\PROJECT\6_D-FLOW_FM\Global Flood Partnership\Example_NigerBasin\Scripts_Tara\Shapefiles\rivwdt_line_10km_clip_Niger_basin_15s_EPSG4326_nodes_data.shp"
    
    path_shapefile_DEM = r"d:\PROJECT\6_D-FLOW_FM\Global Flood Partnership\Example_NigerBasin\Scripts_Tara\Shapefiles\rivwdt_line_10km_clip_Niger_basin_15s_EPSG4326_nodes_DEM_SHEDS_15s.shp"
    
elif path_option == 1:
    
    path_shapefile_points = r"H:\MSc thesis\GIS_Niger_basin\RivWidth\rivwdt_line_10km_clip_Niger_basin_15s_EPSG4326_nodes_data.shp"
    
    path_shapefile_DEM = r"H:\MSc thesis\GIS_Niger_basin\RivWidth\rivwdt_line_10km_clip_Niger_basin_15s_EPSG4326_nodes_DEM_SHEDS_15s.shp"

# <codecell>

# Loading data
shapefile_points_data = ogr.Open(path_shapefile_points)
shapefile_DEM_data    = ogr.Open(path_shapefile_DEM)

# <codecell>

# Assigning all shapefile data to lists/arrays

# reading all data
points_data = shapefile_points_data.GetLayer()
DEM_data    = shapefile_DEM_data.GetLayer()

# checking total points
number_of_points1 = points_data.GetFeatureCount()
number_of_points2 = DEM_data.GetFeatureCount()

if number_of_points1 == number_of_points2:
    number_of_points = number_of_points1
else:
    print('ERROR! shapefiles do NOT have equal number of points!')

# empty arrays
feature_id  = np.zeros(number_of_points)
river_id    = np.zeros(number_of_points)
river_width = np.zeros(number_of_points)
river_depth = np.zeros(number_of_points)
x_coord     = np.zeros(number_of_points)
y_coord     = np.zeros(number_of_points)
DEM_points  = np.zeros(number_of_points)

# filling in arrays, looping over every point in layer
for i in range(number_of_points):
    feature        = points_data.GetFeature(i)
    feature_id[i]  = i
    river_id[i]    = feature.GetField("RIVER_ID")
    river_width[i] = feature.GetField("WIDTH_2")
    river_depth[i] = feature.GetField("DEPTH")
    geometry_temp  = feature.GetGeometryRef()
    x_coord[i]     = geometry_temp.GetX()
    y_coord[i]     = geometry_temp.GetY()
    DEM_points[i]  = (DEM_data.GetFeature(i)).GetField("DEM_SHEDS_")
    
# list with (x,y) coordinates
x_y_coord = zip(x_coord,y_coord)

# <codecell>

# getting junction points from shapefile data

# empty dictionary
D = defaultdict(list)

# fill dictionary with duplicate values and their indices
for i,item in enumerate(x_y_coord):
    D[item].append(i)

D = {k:v for k,v in D.items() if len(v)>1}

# empty list
junction_point_indices = []

# fill in list with indices only
for i in range(len(D)):
    junction_point_indices.append(D[list(D)[i]])
    
# sort list by first index element
junction_point_indices = sorted(junction_point_indices)

# <codecell>

# interpolating river width and depth to all points. 
# trying out option 1: width-based main stream identification

# NOTE: currently using a quick manual interpolation to get a value on river id 2882. Not really nice, as this requires manual checking,
#       but it was a quick, easy fix (and better than what I did before with assiging river id 2883 to all 2882 id's).

# CHOOSE WHETHER TO USE NEXT WIDTH OR AVERAGE WIDTH
# 0 = next, 1 = average
option_check_width = 1

# required library for interpolating
from scipy.interpolate import interp1d

# empty list for saved river id's
saved_river_ids         = []

# empty list for counting indices without values
previous_no_value_index = []

# zero arrays for filling in last two known values/indices
array_two_width_values  = np.zeros(2)
array_two_depth_values  = np.zeros(2)
array_two_values_index  = np.zeros(2)

# adjusting river id at section in main branch without values
# NOTE: turned off, since I am now using manual interpolation to river id 2882
#river_id[list(np.where(river_id==2882)[0])] = 2883

# manual interpolation to a point on river id 2882 (feature id 3164)
river_width[3164] = river_width[939] + 2*((river_width[3173] - river_width[939])/10)
river_depth[3164] = river_depth[939] + 2*((river_depth[3173] - river_depth[939])/10)

# checking river id at first point
current_river_id        = river_id[0]

# first river width/depth with actual value
index_first_river_value = next((i for i, x in enumerate(river_width) if x >=0))

if index_first_river_value > 0:
    
    # assigning same values to all points before this (cannot interpolate from 1 value)
    for i in range(index_first_river_value):
        
        river_width[i]  = river_width[index_first_river_value]
        river_depth[i]  = river_depth[index_first_river_value]

# number of different river id's
unique_river_ids        = list(set(river_id))

# zero array for filling in the order over which river id's the algorithm goes
algorithm_order_river_ids = np.zeros(len(unique_river_ids))

# loop over all different river id's
for i in range(len(unique_river_ids)):
    
    # assign current river id to current index of algorithm order array
    algorithm_order_river_ids[i] = current_river_id
    
    # checking points in current river id
    points_in_current_river_id = sum(river_id==current_river_id)
    
    # if this is the first river id, it might already have some assigned values
    if i != 0:
        start_index = 0
    else:
        start_index = index_first_river_value
    
    # loop over current river id points
    for j in range(start_index, points_in_current_river_id):
        
        # check what would be index in total array
        index_full = np.where(river_id==current_river_id)[0][j]
        
        # check what the river width is at this point
        current_river_width = river_width[index_full]
        
        # check if current point has values
        if (current_river_width >= 0):
            
            # check if interpolation is required
            if len(previous_no_value_index) != 0:
                
                # fill arrays with the two known values
                array_two_width_values[0] = previous_river_width
                array_two_width_values[1] = current_river_width
                
                array_two_depth_values[0] = previous_river_depth
                array_two_depth_values[1] = river_depth[index_full]
                
                # check if there is a junction point within the indices (not counting the first index)
                junction_point_found_in_interpolation = 0
                for k in range(1,len(previous_no_value_index)):
                    
                    if sum(np.array(junction_point_indices)==previous_no_value_index[k]) == 0:    
                        pass
                    else:
                        junction_point_found_in_interpolation = previous_no_value_index[k]
                        previous_no_value_index.remove(junction_point_found_in_interpolation)
                        break
                
                array_two_values_index[1] = len(previous_no_value_index) + 1
                
                # interpolation function
                interpolated_width_function = interp1d(array_two_values_index, array_two_width_values)
                interpolated_depth_function = interp1d(array_two_values_index, array_two_depth_values)
                
                # assign these values to the points with missing values
                for l in range(1,len(previous_no_value_index)+1):
                    
                    river_width[previous_no_value_index[l-1]] = interpolated_width_function(l)
                    river_depth[previous_no_value_index[l-1]] = interpolated_depth_function(l)
                
                # check if junction point was found to assign values to shared points that miss values
                if junction_point_found_in_interpolation == 0:
                    
                    # if not found, pass
                    pass
                
                else:
                    
                    # if found, find all indices at that junction point
                    all_shared_indices = junction_point_indices[np.where(np.array(junction_point_indices) == \
                                                                         junction_point_found_in_interpolation)[0]]
                    
                    # and assign the value of one (which was just interpolated) to all others
                    river_width[all_shared_indices] = river_width[all_shared_indices[np.where(
                                                                                    ~np.isnan(river_width[all_shared_indices]))[0][0]]]
                    
                    river_depth[all_shared_indices] = river_depth[all_shared_indices[np.where(
                                                                                    ~np.isnan(river_depth[all_shared_indices]))[0][0]]]
                
                # reset list for counting indices without values
                previous_no_value_index = []
            
            # assign values at current point to last known values
            previous_river_width = river_width[index_full]
            previous_river_depth = river_depth[index_full]
            
            previous_value_index = index_full
            
        # if not, count indices that are missing values
        else:
            
            previous_no_value_index.append(index_full)
            
        # end point check
        # not an end point if river id of the next point is the same as river id at current point or it is a junction point
        try:
            if (river_id[index_full]==river_id[index_full+1]) or (sum(np.array(junction_point_indices)==index_full) == 1):
                end_point = 0
            else:
                # state as end point
                end_point = 1
                
                # assign last known value to all points with missing values
                for p in range(len(previous_no_value_index)):
                    
                    river_width[previous_no_value_index[p]] = previous_river_width
                    river_depth[previous_no_value_index[p]] = previous_river_depth
                    
                # reset list for counting indices without values
                previous_no_value_index = []
                    
        # but that will give an error if the script is at the last possible point (so index_full +1 will be 'out of bounds')
        except IndexError:
            if (feature_id[-1]!=index_full):
                end_point = 0
            else:
                # state as end point
                end_point = 1
                
                # assign last known value to all points with missing values
                for q in range(len(previous_no_value_index)):
                    
                    river_width[previous_no_value_index[q]] = previous_river_width
                    river_depth[previous_no_value_index[q]] = previous_river_depth
                    
                # reset list for counting indices without values
                previous_no_value_index = []
        
    # check if at junction point to add new river id's to saved list (and equal values at all points of junction point)
    if sum(np.array(junction_point_indices)==index_full) == 1:
        
        # add values of current junction point to other junctions points at same location
        #river_width[junction_point_indices[np.where(np.array(junction_point_indices)==index_full)[0]]] = river_width[index_full]
        #river_depth[junction_point_indices[np.where(np.array(junction_point_indices)==index_full)[0]]] = river_depth[index_full]
        
        # add new river id's to list
        new_river_id_1 = river_id[np.where((x_coord==x_y_coord[index_full][0]) & (y_coord==x_y_coord[index_full][1]))[0][1]]
        new_river_id_2 = river_id[np.where((x_coord==x_y_coord[index_full][0]) & (y_coord==x_y_coord[index_full][1]))[0][2]]
        
        saved_river_ids.append(new_river_id_1)
        saved_river_ids.append(new_river_id_2)
        
    # then, go to the saved river id where known width value is maximum (next or average, based on chosen option)
    if len(saved_river_ids) > 1:
        
        # check known width values at next river id's (if they have a point! otherwise say it is zero)
        try:
            # next width value option
            if option_check_width == 0:
                
                # check if first known value is not at the same junction point
                if (np.where(~np.isnan(river_width[np.where(river_id==new_river_id_1)[0]]))[0][0]) != 0:
                    
                    # assign index of first known value
                    index_first_known_value_1 = np.where(~np.isnan(river_width[np.where(river_id==new_river_id_1)[0]]))[0][0]
                    
                    river_width_1             = river_width[np.where(river_id==new_river_id_1)[0][index_first_known_value_1]]
                    
                # otherwise, first known value is at junction point (and comparing new branches would be impossible)
                else:
                    
                    # assign index of first known value after junction point (second known value)
                    index_first_known_value_1 = np.where(~np.isnan(river_width[np.where(river_id==new_river_id_1)[0]]))[0][1]
                    
                    river_width_1             = river_width[np.where(river_id==new_river_id_1)[0][index_first_known_value_1]]
            
            # average width value option
            elif option_check_width == 1:
                
                # check if river id has any points with a value
                if sum(~np.isnan(river_width[np.where(river_id==new_river_id_1)[0]])) > 0:
                    
                    river_width_1 = np.mean(river_width[np.where(river_id==new_river_id_1)[0]][np.where(
                                                                    ~np.isnan(river_width[np.where(river_id==new_river_id_1)[0]]))])
                    
                # otherwise, assign a value of zero
                else:
                    
                    river_width_1 = 0
                
        # if no value can be found, assign a value of zero
        except:
            river_width_1 = 0
        
        # same procedure for newly found river id 2
        try:
            if option_check_width == 0:
                if (np.where(~np.isnan(river_width[np.where(river_id==new_river_id_2)[0]]))[0][0]) != 0:
                    index_first_known_value_2 = np.where(~np.isnan(river_width[np.where(river_id==new_river_id_2)[0]]))[0][0]
                    river_width_2             = river_width[np.where(river_id==new_river_id_2)[0][index_first_known_value_2]]
                else:
                    index_first_known_value_2 = np.where(~np.isnan(river_width[np.where(river_id==new_river_id_2)[0]]))[0][1]
                    river_width_2             = river_width[np.where(river_id==new_river_id_2)[0][index_first_known_value_2]]
            elif option_check_width == 1:
                if sum(~np.isnan(river_width[np.where(river_id==new_river_id_2)[0]])) > 0:
                    river_width_2 = np.mean(river_width[np.where(river_id==new_river_id_2)[0]][np.where(
                                                                    ~np.isnan(river_width[np.where(river_id==new_river_id_2)[0]]))])
                else:
                    river_width_2 = 0
        except:
            river_width_2 = 0
        
        # take the river id with maximum width at its first known point
        if river_width_1 >= river_width_2:
            
            current_river_id = saved_river_ids[-2]
            
            # and remove this from the saved list
            del(saved_river_ids[-2])
            
        else:
            
            current_river_id = saved_river_ids[-1]
            
            # and remove this from the saved list
            del(saved_river_ids[-1])
        
    # or, if only 1 river id is left, just use this one
    elif len(saved_river_ids) != 0:
        
        current_river_id = saved_river_ids[-1]
        
        # and remove this from the saved list
        del(saved_river_ids[-1])

    # check if there are still points without value
    # NOTE1: only required because I changed river id 2882 to 2883 (at 'adjusting river id at section in main branch without values')
    #        with this change, the code goes past the junction point that links river id's 2882 (no longer exists), 2883 and 2930
    #        and since the main branch goes over 2883, river id 2930 is never handled and keeps nan values
    #        I thought this was the nicest way of handling this issue, instead of just doing something specifically for id 2930
    # NOTE2: turned off, since I am now using manual interpolation to river id 2882
    #elif len(np.where(np.isnan(river_width))[0]) > 0:
        
        #current_river_id = river_id[np.where(np.isnan(river_width))[0]][0]
        
    # if last point was an end point, assign values of junction point of new river id
    if end_point == 0:
        
        pass
    
    else:
        
        # index of new starting point
        new_starting_point   = np.where(river_id==current_river_id)[0][0]
        
        # find indices that are at the same location of new starting point
        shared_indices       = np.where((x_coord==x_y_coord[new_starting_point][0]) & (y_coord==x_y_coord[new_starting_point][1]))[0]
        
        # find the index with values
        previous_value_index = shared_indices[np.where(~np.isnan(river_width[shared_indices]))[0][0]]
        
        # set values at this point as previously known values
        previous_river_width = river_width[previous_value_index]
        previous_river_depth = river_depth[previous_value_index]
            
        # and set all values that share this location equal to those of the point that has values
        for r in range(len(shared_indices)):
            river_width[shared_indices[r]] = previous_river_width
            river_depth[shared_indices[r]] = previous_river_depth

# <codecell>

# test to see if there are any nan values left
print(len(np.where(np.isnan(river_width))[0]))
print(len(np.where(np.isnan(river_depth))[0]))
print(len(np.where(np.isnan(DEM_points))[0]))

# <markdowncell>

# Ok, now every point has a width and depth. Every point also has an elevation through the DEM. My original idea was to burn the DEM based on the depth of the river, but now I think that I should just "smooth" the DEM and have the depths create the "burning" implicitely in the 1D network.

# <codecell>

# checking if I can create a new array to keep the DEM data array unchanged

# this works:
elevation_values = np.zeros(len(DEM_points))
for i in range(len(DEM_points)):
    elevation_values[i] = DEM_points[i]
elevation_values[0] = 3
print(elevation_values[0])
print(DEM_points[0])
print('')

# this also works:
elevation_values = np.zeros(len(DEM_points))
for i in range(len(DEM_points)):
    elevation_values[i] = DEM_points[i]
elevation_values[0] = elevation_values[1]
elevation_values[1] = 3
print(elevation_values[0])
print(elevation_values[1])

# this does not:
#elevation_values = DEM_points
#elevation_values[0] = 3
#print(elevation_values[0])
#print(DEM_points[0])

# <markdowncell>
```
