```python
import netCDF4
import matplotlib.pyplot as plt
```

```python
frad = '../static/data/drad250.1grid.ICE5Gv1.3_VM2_L90_2012.nc'
fsea = '../static/data/dsea250.1grid.ICE5Gv1.3_VM2_L90_2012.nc'
fgeo = '../static/data/dGeoid250.1grid.ICE5Gv1.3_VM2_L90_2012.nc'
```

```python
fig, axes= plt.subplots(1,3, figsize=(15,4))
names = ['DGeoid_250', 'Drad', 'Dsea_250']
files = [fgeo, frad, fsea]

for (fname, varname, ax) in zip(files, names, axes):
    ds = netCDF4.Dataset(fname)
    V = ds.variables[varname][:]
    lat = ds.variables['Lat'][:]
    lon =ds.variables['Lon'][:]
    ds.close()

    im = ax.pcolorfast(lon, lat, V)
    ax.set_title(varname)
    plt.colorbar(im, ax=ax)
```
