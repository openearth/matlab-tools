```python
import matplotlib.pyplot as plt
import netCDF4
import pandas
import bisect

#plt.xkcd()
```

```python
ds_u = netCDF4.Dataset('uwnd.10m.mon.mean.nc')
lat = ds_u.variables['lat'][:]
lon = ds_u.variables['lon'][:]
t = netCDF4.num2date(ds_u.variables['time'],ds_u.variables['time'].units)
u = ds_u.variables['uwnd'][:,:,:]
u_mean = ds_u.variables['uwnd'][:,:,:].mean(0)
ds_u.close()
ds_v = netCDF4.Dataset('vwnd.10m.mon.mean.nc')
lat = ds_v.variables['lat'][:]
lon = ds_v.variables['lon'][:]
t = netCDF4.num2date(ds_v.variables['time'],ds_v.variables['time'].units)
v = ds_v.variables['vwnd'][:,:,:]
v_mean = ds_v.variables['vwnd'][:,:,:].mean(0)
ds_v.close()
```

```python
fig, ax = plt.subplots(figsize=(30,14))
us = np.sqrt(u_mean**2+v_mean**2)
lon_idx = lon < 10
lat_idx = np.logical_and(lat >= 50, lat < 60)
im = ax.pcolorfast(lon[lon_idx], lat[lat_idx], us[lon_idx,:][:,lat_idx], cmap='Greens')
s = ax.streamplot(lon[lon_idx], lat[lat_idx], u_mean[lon_idx,:][:,lat_idx], v_mean[lon_idx,:][:,lat_idx], linewidth=us[lon_idx,:][:,lat_idx], color=us[lon_idx,:][:,lat_idx], cmap='Oranges')
plt.colorbar(im, ax=ax)
```

```python
lon_i = bisect.bisect(lon, 3)
lat_i = bisect.bisect(-lat, -52) # lat is reversed
fig, ax = plt.subplots(figsize=(20,5))
u_i = u[:,lat_i, lon_i]
v_i = v[:,lat_i, lon_i]
l, = ax.plot(t, np.sqrt(u_i**2+v_i**2))
l.set_lod(True)
ax.minorticks_on()
```

```python
num = 300
lat_i = np.linspace(0,np.pi, num=num, endpoint=False)
lon_i = np.mod(np.linspace(0,np.pi*2, num=num, endpoint=False), np.pi*2)
fig, axes = plt.subplots(1,4, figsize=(20,4))
for i in range(4):
    F_u = scipy.interpolate.RectSphereBivariateSpline(np.deg2rad(lat[::-1]) + 0.5*np.pi, np.deg2rad(lon), u[-i,::-1,:])
    F_v = scipy.interpolate.RectSphereBivariateSpline(np.deg2rad(lat[::-1]) + 0.5*np.pi, np.deg2rad(lon), v[-i,::-1,:])
    ax = axes[i]
    ax.pcolorfast(np.rad2deg(lon_i), np.rad2deg(lat_i ), np.sqrt(F_u(lat_i, lon_i)**2 + F_v(lat_i, lon_i)**2), cmap='Blues')
    ax.streamplot(np.rad2deg(lon_i), np.rad2deg(lat_i ), F_u(lat_i, lon_i), F_v(lat_i, lon_i), linewidth=0.3*np.sqrt(F_u(lat_i, lon_i)**2 + F_v(lat_i, lon_i)**2))
```
