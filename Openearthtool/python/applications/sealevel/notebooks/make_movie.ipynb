```python
dst_srs = osgeo.osr.SpatialReference()
dst_srs.ImportFromEPSGA(900913)
src_srs = osgeo.osr.SpatialReference()
src_srs.ImportFromEPSG(4326)

dst2src = osgeo.osr.CoordinateTransformation(dst_srs, src_srs)
src2dst = osgeo.osr.CoordinateTransformation(src_srs, dst_srs)
```

```python
# at which resolution do we want to generate a movie
width = lon.shape[0]
height = lat.shape[0]
```

```python
# the extent of the 900913 coordinate system
dst_xmin, dst_ymin = -20037508.34, -20037508.34
dst_xmax, dst_ymax = 20037508.34, 20037508.34
lonmin, latmin, _ = dst2src.TransformPoint(dst_xmin, dst_ymin)
lonmax, latmax, _ = dst2src.TransformPoint(dst_xmax, dst_ymax)
```

```python
# transform into a matrix of lat/lon per image point
x, y = np.ogrid[dst_xmin:dst_xmax:(width*1j), dst_ymin:dst_ymax:(height*1j)]
dst_lon = [x[0] for x in dst2src.TransformPoints(np.c_[x, np.zeros_like(x)])]
dst_lat = [y[1] for y in dst2src.TransformPoints(np.c_[np.zeros_like(y.squeeze())+x[0], y.squeeze()])]
```

```python
# this is why we're doing so difficult, coordinate systems do not match up
plt.plot(dst_lat, np.gradient(dst_lat), label='gradient of destination')
plt.plot(lat[:], np.gradient(lat[:]), label='gradient of source')
plt.legend(loc='best')
```

```python
# We can use a simple linear interpolation to match these up
L = scipy.interpolate.interp1d(lat[:], d[0], bounds_error=False)
L.y[:] = d[1]
V = L(dst_lat)
#V = np.ma.masked_array(V, np.logical_or.reduce([np.isnan(V), V < -1000, V>1000]))
#V.fill_value = 0
plt.figure(figsize=(20,15))
plt.imshow(V.T, vmin=-50, vmax=50, origin='lower', cmap='ocean')
plt.colorbar()
```

```python
N = matplotlib.colors.Normalize(vmin=-70, vmax=70)
C = lambda x: matplotlib.cm.ocean(x, bytes=True)

#for i in range(20):
for i in range(d.shape[0]):
    d_i = d[i]
    L.y[:] = d_i
    values = L(dst_lat).T
    # mask
    mask = np.logical_or.reduce([np.isnan(values), values[:,:] < -1000, values[:,:] > 1000])
    values[mask] = 0
    # convert to colors
    imgcolor = C(N(np.flipud(values)))[:,:,:3]
    # convert to bytes
    imgmask = np.zeros_like(imgcolor)
    # set to white
    imgmask[np.flipud(mask),:] = 255
    # Blur
    imgmaskpil = PIL.Image.fromarray(imgmask)
    imgmaskpil = imgmaskpil.filter(PIL.ImageFilter.GaussianBlur(radius=5))
    # Overwrite
    imgmask = np.array(imgmaskpil)
    # Combine
    img = np.concatenate((imgcolor, imgmask), axis=0)
    # split and merge
    img = np.concatenate([img[:,540:,:],img[:,:540,:]],axis=1)
    plt.imsave('test_{:03d}.png'.format(i), img)
values, mask
```
