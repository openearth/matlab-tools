```python
import netCDF4
import matplotlib.pyplot as plt
```

```python
url = 'http://iridl.ldeo.columbia.edu/SOURCES/.Sandwell/.seafloor/.height/dods'
thin = 20
```

```python
ds = netCDF4.Dataset(url)

X = ds.variables['X'][::thin]
Y = ds.variables['Y'][::thin]
H = ds.variables['height'][::thin,::thin]
ds.close()
```

```python
fig, ax = plt.subplots(1,1, figsize=(20,10))
split = 500
vmax = 5000
mesh = ax.pcolormesh(X, Y, H, cmap=matplotlib.cm.terrain, vmin=-0.25*vmax,vmax=vmax)
plt.colorbar(mesh, ax=ax)
mesh = ax.pcolormesh(X-360, Y, H, cmap=matplotlib.cm.gist_earth, vmin=-vmax,vmax=vmax)
ax.autoscale()
plt.colorbar(mesh, ax=ax)
```
