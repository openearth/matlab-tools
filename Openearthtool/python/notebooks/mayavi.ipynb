```python

def stream2vorticity(st):
    start = 0
    vorticities = []
    for i in range(st.output.lines.number_of_cells):
        """loop over all lines"""
        lines = st.output.lines.data.to_array()
        vorticity = st.output.point_data.get_array(6).to_array()
        n = lines[start]
        idx = lines[start+1:start+n]
        vorticities.append(vorticity[idx])
        start += (n + 1)
    return np.asarray(vorticities)

def stream2points(st):
    start = 0
    pointlist = []
    for i in range(st.output.lines.number_of_cells):
        """loop over all lines"""
        lines = st.output.lines.data.to_array()
        points = st.output.points.to_array()
        n = int(lines[start])
        idx = int(lines[start+n])
        point = points[idx]
        pointlist.append(point)
        start += (n + 1)
    return np.array(pointlist)
    
    
```
