```python
from ecmwfapi import ECMWFDataServer
import matplotlib.pyplot as plt
```

```python
server = ECMWFDataServer()
 
server.retrieve({
    'dataset' : "interim",
    'step'    : "0",
    'date'    : "20130101/to/20130102",
    'time'    : "00",
    'levtype' : "sfc",
    'class'   : "ei",
    'grid'    : "0.75/0.75",
    'target'  : r'd:\Scheel\Documents\notebooks\data\data.nc'
    })
```

```python
import netCDF4

ds = netCDF4.Dataset('p:\Archive-Projects\MetoceanData\open\ERAinterim\data\erai2012.nc')

ds.variables.keys()
```

```python
long = ds.variables['longitude'][:]
latt = ds.variables['latitude'][:]

plotvars = set(ds.variables.keys()) - {"longitude", "latitude", "time"}
plotvars
```

```python
fig, axes = plt.subplots(1, len(plotvars), sharey=True, figsize=(25,3))
for varname, ax in zip(plotvars, axes):
    var = ds.variables[varname]
    im = ax.pcolormesh(long,latt,var[0,:,:], cmap='Blues')
    plt.colorbar(im, ax=ax)
    ax.set_title(var.long_name)
# long.shape

# long[:]
```
