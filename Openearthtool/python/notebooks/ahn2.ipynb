```python
import osgeo.ogr
```

```python
osgeo.ogr.RegisterAll()
drivers = {osgeo.ogr.GetDriver(i).GetName():osgeo.ogr.GetDriver(i) for i in range(osgeo.ogr.GetDriverCount())}
```

```python
driver = drivers['WFS']
```

```python
ds = driver.Open('WFS:' + url)
layer = ds.GetLayer(0)
# why are features 1 based?
features = [layer.GetFeature(i+1) for i in range(layer.GetFeatureCount())]
```

```python
rows = []
for feature in features:
    row = {feature.GetFieldDefnRef(i).GetName(): feature.GetField(i) for i in range(feature.GetFieldCount())}
    rows.append(row)
```

```python
import pandas
df = pandas.DataFrame.from_records(rows)
df.head()
```

```python
df.plot('lo_x', 'lo_y', style='k.')
```

```python
import liblas
```

```python
las = liblas.file.File('/home/fedor/data/ahn2/g30dz1.laz')
```

```python
header = las.get_header()
header.get_count()

def las2rows(las, step=10000):
    for i in xrange(0, header.get_count(), step):
        point = las.read(i) 
        color = point.get_color()
        
        yield point.x, point.y, point.z, (color.red, color.green, color.blue)
df = pandas.DataFrame.from_records(las2rows(las), columns=['x', 'y', 'z', 'color'])
df.head()
```

```python
df.shape, header.get_count()
```

```python
plt.scatter(df['x'], 
            df['y'], 
            c=df['z'], 
            marker='.', cmap='gist_earth', vmax=5, vmin=-5)
```
